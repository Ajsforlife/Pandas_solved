```python
# Dependencie
import pandas as pd
```

```python
# Load in file
movie_file = "Resources/movie_scores.csv"
```

```python
# Read and display the CSV with Pandas
movie_scores = pd.read_csv(movie_file)
movie_scores
```

```python
# List all the columns in the table
print(movie_scores.columns)
```

```python
# We only want IMDb data, so create a new table that takes the Film and all the columns relating to IMDB
movie_scores_small = movie_scores[['FILM','IMDB','IMDB_norm','IMDB_norm_round','IMDB_user_vote_count']].copy()
movie_scores_small
```

```python
# We only like good movies, so find those that scored over 7, and ignore the norm rating
good_movies = movie_scores[(movie_scores["IMDB"] >= 7)]
good_movies
```

```python
# Find less popular movies--i.e., those with fewer than 20K votes
unpopular_movies = movie_scores[(movie_scores["IMDB_user_vote_count"] <= 20000)]
unpopular_movies
```
